```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier 
```

```python
df.columns
```

```python
X=df.loc[:,['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
       'exang', 'oldpeak', 'slope', 'ca', 'thal']]
y=df.loc[:,'target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1429)
```

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=1429, ratio=1.0)
X_train, y_train = sm.fit_sample(X_train, y_train)
```

```python
rf=RandomForestClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
hyper={"max_depth":[2.3,4],"max_features":['auto','sqrt','log2',None], 
       "criterion":['gini','entropy'], 
       "bootstrap":[True,False]}
gdcv=GridSearchCV(estimator=rf,param_grid=hyper, cv=10, n_jobs=4,scoring='accuracy', return_train_score=True)
bestF = gdcv.fit(X_train, y_train)
```

```python
print('Best Parameters using grid search: \n', 
      bestF.best_estimator_)
```

```python
rf=RandomForestClassifier(bootstrap=True, class_weight=None, criterion='entropy',
                       max_depth=4, max_features='sqrt', max_leaf_nodes=None,
                       min_impurity_decrease=0.0, min_impurity_split=None,
                       min_samples_leaf=1, min_samples_split=2,
                       min_weight_fraction_leaf=0.0, n_estimators=10,
                       n_jobs=None, oob_score=False, random_state=None,
                       verbose=0, warm_start=False)
```

```python
rf.fit(X_train, y_train)
```

```python
pred1=rf.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(y_test, pred1))
```

```python
Accuracy=metrics.accuracy_score(y_test, pred1)
print("Accuracy is ",np.round(Accuracy,2))
```

```python
AUC=metrics.roc_auc_score(y_test, pred1)
print("AUC is ",np.round(AUC,2))
```

```python
fpr, tpr, th = metrics.roc_curve(y_test,rf.predict_proba(X_test)[:,1])
```

```python
sns.lineplot(fpr, tpr, ci=None)
sns.lineplot([0,1], [0,1])
plt.legend(title="AUC", labels=[np.round(AUC,2)], loc='lower right')
```

```python
cm=metrics.confusion_matrix(y_test, pred1)
ax=sns.heatmap(cm,annot=True)
ax.set_ylim(2,0)
```
